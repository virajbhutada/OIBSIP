# Task 2: Unemployment Analysis with Python


Unemployment is measured by the unemployment rate which is the number of people
who are unemployed as a percentage of the total labour force. We have seen a sharp
increase in the unemployment rate during Covid-19, so analyzing the unemployment rate
can be a good data science project.





- **Objective**: Analyzing the impact of the Covid-19 pandemic on unemployment rates.
  
- **Measurement**: Unemployment rates represent the percentage of unemployed individuals in the total labor force.
  
- **Techniques Used**:
  - **Python**: The primary programming language used for data analysis and visualization.
  - **Data Visualization Libraries**:
    - **Matplotlib**: Utilized for creating static, high-quality visualizations like box plots and bar charts.
    - **Seaborn**: Employed for statistical data visualization, including heatmaps and count plots.
    - **Plotly Express**: Used for interactive visualizations, such as sunburst plots and animated scatter plots.
  
- **Visualizations**:
  - **Count Plots**: Visualized regional and area distributions to identify variations in unemployment rates.
  - **Heatmaps**: Analyzed correlations between economic indicators, revealing patterns among unemployment rate, employment, and labor participation rate.
  - **Sunburst Plots**: Provided a hierarchical view of unemployment rates across different areas and regions.
  - **Animated Scatter Plots**: Illustrated temporal changes, offering insights into the evolution of unemployment rates during different months.
  
- **Insights**:
  - **Regional Disparities**: Identified variations in unemployment rates across different regions, highlighting areas of concern.
  - **Correlations**: Discovered negative correlations between employment and unemployment, aiding in understanding economic dynamics.
  - **Temporal Trends**: Analyzed how unemployment rates changed over time, crucial for recognizing pandemic-related impacts.

- **Significance**:
  - **Policymakers**: Informed decision-making regarding targeted interventions and policy adjustments.
  - **Businesses**: Strategic planning for adapting to economic challenges and understanding employment trends for workforce management.
  
- **Outcome**: This analysis equips stakeholders with actionable insights, enabling effective responses to the increased unemployment rates during the Covid-19 crisis.

# Importing Necessary Libraries

```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pl
```

# Loading and Exploring the Dataset

```python
df = pd.read_csv("Unemployment in India.csv")
```

```python
# Removing spaces before and after the column names
df.columns = df.columns.str.strip()
df
```

```python
# Display the first few rows of the dataset to understand its structure
df.head()
```

# Data Cleaning and Preprocessing

```python
# Check for missing values in the dataset and handle them if necessary
df.isnull().sum()
```

```python
# Removing the null values in the dataset
df = df.dropna()
df
```

```python
# Getting the columns alone
df.columns
```

```python
# Checking the size of the dataset
df.size
```

```python
# Checking the shape of the dataset
df.shape
```

```python
# Checking the data types of the dataset
df.dtypes
```

```python
# Getting info of the dataset
df.info()
```

```python
# Checking for duplicate records in the dataset
df.duplicated().sum()
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Descriptive statistical analysis of data
df.describe()
```

```python
# Counting the values of regions
df['Region'].value_counts()
```

```python
# Counting the values of dates
df['Date'].value_counts()
```

```python
# Counting the values of areas
df['Area'].value_counts()
```

# Creating Additional Columns in the DataFrame

```python
# Adding Day, Month, and Year
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year

# Importing Warnings library
import warnings

# Ignoring unwanted warnings
warnings.filterwarnings("ignore")

df
```

# **Graphical Representation of Data**

# Count Plot of the Regions

```python
# Count Plot of the regions
sns.countplot(x = df['Region'], palette = "bright")
plt.xticks(rotation ='vertical')
plt.show()
```

# Count Plot of the Areas


```python
# Count Plot of the Areas
plt.figure(figsize = (3, 2))
sns.countplot(x = df['Area'],palette = "bright")
plt.xticks(rotation='vertical')
plt.show()
```

# Heatmap of Estimated Unemployment Rate, Estimated Employed, and Estimated Labour Participation Rate

```python
# Heatmap of Estimated Unemployement Rate, Estimated Employed, and Estimated Labour Participation Rate
plt.figure(figsize = (3,2))
sns.heatmap(df[['Estimated Unemployment Rate (%)',
                'Estimated Employed',
                'Estimated Labour Participation Rate (%)']].corr(),annot=True)
plt.show()
```

# Sunburst Plot of Unemployment Rate

```python
# Sunburst plot of Unemplotement Rate
import plotly.express as px  # Import the px module from Plotly

unemploment = df[["Area", "Region", "Estimated Unemployment Rate (%)"]]
fig = px.sunburst(unemploment, path = ["Area", "Region"],
                     values = "Estimated Unemployment Rate (%)",
                     width = 700, height=700,
                     title ="Unemployment Rate")
fig.show()

```

# Pie Chart of the Area (Rural and Urban)

```python
# Pie Chart of the Area (Rural and Urban)
plt.pie(x = df['Area'].value_counts(), autopct = '%0.2f', explode = [0,0.09], colors = [ '#d3d3d3','#f5deb3'], labels = ['Urban', 'Rural'])
plt.show()
```

# Box Plot of Estimated Unemployment Rate

```python
# Box Plot of Estimated Unemployment Rate
plt.figure(figsize=(24, 7))
sns.boxplot(x = 'Region', y= 'Estimated Unemployment Rate (%)', data = df)
plt.xticks(rotation = 'vertical')
plt.show()
```

# Bar Plot of Average Estimated Unemployment Rate and Estimated Employed People in Every Month

```python
# Bar Plot of Average Estimated Unployment Rate
fig = px.bar(df, x = 'Region', y = "Estimated Unemployment Rate (%)", color = "Region", title = "Average unemploment Rate")
fig.update_layout(xaxis = {'categoryorder':'total descending'})
fig.show()

# Bar Plot of Estimated Employed people in every month
fig = px.bar(df, x = 'Month', y = 'Estimated Employed', color = 'Month', title = 'Estimated Employed People')
fig.show()
```

# Scatter Plot Matrix

```python
fig = px.scatter_matrix(df, dimensions=['Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)'], color='Region')

fig.show()
```

# Scatter Plot of Estimated Employed in Animation

```python
# Scatter Plot of Estimated Employed in animation
pl.scatter(data_frame = df, x = 'Estimated Employed', y = 'Region',animation_frame = 'Month')
```

# Scatter Plot of Estimated Unemployment Rate in Animation

```python
# Scatter Plot of Estimated Unemployment Rate in animation
pl.scatter(data_frame = df, x = 'Estimated Unemployment Rate (%)', y = 'Region', animation_frame = 'Month')
```

# Scatter Plot of Estimated Labour Participation Rate in Animation

```python
# Scatter Plot of Estimated Labour Participation Rate in animation
pl.scatter(data_frame = df, x = 'Estimated Labour Participation Rate (%)', y = 'Region', animation_frame = 'Month')
```
